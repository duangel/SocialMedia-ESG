```python
import pandas as pd
import os
import glob
```

Code to pre-process Instagram data, turning any user tags or links to generic @user and http

```python
def replaceInsta(strings):
  '''
  Generalize user tags and links
  '''
    processed = []
    for s in strings:
        try:
          w = s.split(' ')
        except:
          s = str(s)
          w = s.split(' ')

        allusers = list(filter(lambda x: '@' in x, w))
        allusersLink = list(filter(lambda x: 'https' in x, w))

        mapping = []
        mappingLink = []

        for a in allusers:
          mapping.append((a,'@user'))

        for a in allusersLink:
          mappingLink.append((a,'https'))

        for k, v in mapping:
            w = list(map(lambda x: x.replace(k,v), w))

        for k, v in mappingLink:
            w = list(map(lambda x: x.replace(k,v), w))

        processed.append(' '.join(w))
    return processed

## Begin Instagram preprocessing
filelist = sorted(glob.glob('./data/*.csv'))

path = r'./data'

error = []
for file in filelist:

    print("--------------- Starting with %s ---------------"%file[7:])
    compDf = pd.read_csv(file)
    allPostDescriptions = compDf.Description.tolist()

    try:
        allPostComments = compDf.PostsComments.tolist()
    except:
        error.append(os.fsdecode(file))
        continue

    compDf['Description'] = replaceInsta(allPostDescriptions)

    compDf['PostsComments'] = replaceInsta(allPostComments)

    if ('Unnamed: 0' in compDf.columns) and ('PostComments' in compDf.columns):
      compDf.drop(['Unnamed: 0', 'PostComments'], axis = 1, inplace = True)

    if 'PostComments' in compDf.columns:
      compDf.drop(['PostComments'], axis = 1, inplace = True)

    if 'Unnamed: 0' in compDf.columns:
      compDf.drop(['Unnamed: 0'], axis = 1, inplace = True)

    compDf.to_csv(file)

    print("--------------- Done with %s ---------------"%file[7:])
```
